```python
# prompt: calculate the metrices for each dataset in the datasetcolumn

from sklearn.metrics import confusion_matrix

# Function to compute Quadratic Weighted Kappa (QWK)
def quadratic_weighted_kappa(y_true, y_pred):
    # Compute confusion matrix
    conf_matrix = confusion_matrix(y_true, y_pred)

    # Compute the weight matrix
    num_classes = len(np.unique(y_true))
    weight_matrix = np.zeros((num_classes, num_classes))
    for i in range(num_classes):
        for j in range(num_classes):
            weight_matrix[i, j] = (i - j) ** 2 / (num_classes - 1) ** 2

    # Normalize the confusion matrix
    total_weight = np.sum(weight_matrix * conf_matrix)
    total_weight_sum = np.sum(conf_matrix) * (num_classes - 1) ** 2

    # Calculate QWK score
    kappa_score = 1 - (total_weight / total_weight_sum)
    return kappa_score

# Assuming you have your dataset column named 'datasetcolumn' in a DataFrame 'df'
for dataset in df_test['dataset'].unique():
    # Filter data for the current dataset
    subset = df_test[df_test['dataset'] == dataset]

    # Assuming 'y_true' and 'y_pred' are your true and predicted labels for the subset
    y_true = subset['score']  # Replace 'y_true' with your actual column name
    y_pred = subset['y_pred']  # Replace 'y_pred' with your actual column name

    # Calculate metrics
    mse = mean_squared_error(y_true, y_pred)
    rmse=root_mean_squared_error(y_true, y_pred)
    mae=mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred.round())
    precision = precision_score(y_true, y_pred.round(), average='weighted')
    recall = recall_score(y_true, y_pred.round(), average='weighted')
    f1 = f1_score(y_true, y_pred.round(), average='weighted')
    kappa = cohen_kappa_score(y_true, y_pred.round())

    # Convert y_pred to discrete class labels for QWK
    y_pred_classes = np.round(y_pred).astype(int)
    qwk = quadratic_weighted_kappa(y_true, y_pred_classes)
    qwk2=cohen_kappa_score(y_true, y_pred, weights='quadratic')
    persorn=np.corrcoef(y_true, y_pred)[0, 1]
    print(f"Metrics for dataset: {dataset}")
    print(f"Mean Squared Error: {mse}")
    print(f"Root Mean Squared Error: {rmse}")
    print(f"Mean Absolute Error: {mae}")
    print(f"R-squared: {r2}")
    print(f"Accuracy: {accuracy}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1-score: {f1}")
    print(f"Cohen's Kappa: {kappa}")
    print(f"Quadratic Weighted Kappa: {qwk}")
    print(f"Quadratic Weighted Kappa2: {qwk2}")
    print(f"Pearson: {persorn}")
    print("-" * 20)

```
